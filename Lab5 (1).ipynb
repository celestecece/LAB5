##Discussion: In this lab, we learned to recognize and differenciate the basic EEG patterns. To gain an understanding of the clinical applications of EEGs. The variables and understanding the factors that might influence the time required to complete a reflex.


The distinction between monosynaptic and polysynaptic reflexes lies in the number of synapses involved in the reflex arc, which is the neural pathway mediating a reflex action. 

Monosynaptic Reflex:

In a monosynaptic reflex, there is only one synapse between the sensory (afferent) neuron and the motor (efferent) neuron.
The sensory neuron directly stimulates the motor neuron, leading to a rapid and simple response.
A monosynaptic reflex is the knee-jerk reflex (patellar reflex), where the stretch of the quadriceps muscle activates a sensory neuron, leading to a quick contraction of the muscle.

Polysynaptic Reflex:

In a polysynaptic reflex, there are multiple synapses involving one or more interneurons (association neurons) between the sensory and motor neurons.
The interneurons serve to integrate and process the incoming sensory information, allowing for more complex and modifiable responses.
A polysynaptic reflex is the withdrawal reflex, where the sensory neuron activates interneurons in the spinal cord, leading to the contraction of muscles to withdraw from a painful stimulus.
